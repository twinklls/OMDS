# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

## My answers

1. Improvements for Nitrate & Phosphate Line Graphs
- Using a dual-axis chart instead of separate graphs → Easier comparison.
- Using smoothing techniques (rolling average) to reduce fluctuations.
- Changing line styles (solid vs. dashed) & colors for differentiation.
- Annotating key points to highlight trends.
- Using a light grid & remove excessive axis labels for clarity.

2. Enhancements for Stacked Area Chart
- Using meaningful labels instead of "ser1" & "ser2".
- Making colors distinguishable yet muted.
- Reducing opacity (alpha) to avoid excessive overlap.
- Adding a title & refine the x-axis labels for better readability.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numdates = 12
np.random.seed(0)
datearray = pd.date_range(start="2023/01/01", end="2024/01/01", freq="M")

# Extracted nitrate & phosphate levels
nitrate_levels = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)

np.random.seed(1)
phosphate_levels = 100 * (1 - (np.random.normal(size=numdates) / 8 + np.arange(numdates) / 16))

# ----- Enhanced Dual Y-Axis Line Graph -----
print("\nDisplaying Enhanced Dual-Axis Graph: Nitrate & Phosphate Levels")

fig, ax1 = plt.subplots(figsize=(10, 5))

sns.set_style("whitegrid")  # Minimalist grid for readability

# Nitrate line (Primary Y-Axis)
ax1.plot(datearray, nitrate_levels, color="blue", linestyle="solid", marker="o", alpha=0.7, label="Nitrate Level")
ax1.set_xlabel("Year and Month", fontsize=12)
ax1.set_ylabel("Nitrate Level (% of Eutrophic Value)", color="blue", fontsize=12)

# Add a second Y-axis for Phosphate
ax2 = ax1.twinx()
ax2.plot(datearray, phosphate_levels, color="red", linestyle="dashed", marker="s", alpha=0.7, label="Phosphate Level")
ax2.set_ylabel("Phosphate Level (% of Eutrophic Value)", color="red", fontsize=12)

# Annotations for key points
ax1.annotate("Nitrate Spike", xy=(datearray[-1], nitrate_levels[-1]), xytext=(-50, -20), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="blue"), fontsize=10, color="blue")
ax2.annotate("Phosphate Dip", xy=(datearray[-2], phosphate_levels[-2]), xytext=(-50, 10), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="red"), fontsize=10, color="red")

# Hide top spines for a cleaner look
ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)

plt.title("Enhanced Nitrate & Phosphate Level Trends", fontsize=14)
plt.show()
```

```python
# ----- Enhanced Stacked Area Chart -----
print("\nDisplaying Enhanced Stacked Area Chart")

# DataFrame for plotting
df = pd.DataFrame({"Nitrate Level": nitrate_levels, "Phosphate Level": phosphate_levels}, index=datearray)

# Stacked area chart
fig, ax = plt.subplots(figsize=(10, 5))
df.plot.area(ax=ax, alpha=0.7, colormap="coolwarm")

plt.xlabel("Year and Month", fontsize=12)
plt.ylabel("Level (% of Eutrophic Value)", fontsize=12)
plt.title("Enhanced Stacked Area Chart: Nitrate & Phosphate", fontsize=14)
plt.legend(title="Water Quality Indicators", fontsize=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset_path = "dataset/austinHousingData.csv"  # Update path if necessary
df = pd.read_csv(dataset_path)

# Convert relevant columns to numeric and handle missing values
df["propertyTaxRate"] = pd.to_numeric(df["propertyTaxRate"], errors="coerce")
df["avgSchoolRating"] = pd.to_numeric(df["avgSchoolRating"], errors="coerce")

# Drop rows with missing values in these columns
df_cleaned = df.dropna(subset=["propertyTaxRate", "avgSchoolRating"])

# Ensure values are numeric
df_cleaned = df_cleaned[(df_cleaned["propertyTaxRate"].apply(lambda x: isinstance(x, (int, float)))) &
                        (df_cleaned["avgSchoolRating"].apply(lambda x: isinstance(x, (int, float))))]

# ----- Line Plot with Dual Y-Axis -----
print("\nDisplaying Line Plot: Property Tax Rate vs. Average School Rating")

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.set_xlabel("Index (Properties)")
ax1.set_ylabel("Property Tax Rate", color="blue")
ax1.plot(df_cleaned.index, df_cleaned["propertyTaxRate"], color="blue", alpha=0.6, label="Property Tax Rate")

ax2 = ax1.twinx()
ax2.set_ylabel("Average School Rating", color="orange")
ax2.plot(df_cleaned.index, df_cleaned["avgSchoolRating"], color="orange", alpha=0.6, label="Average School Rating")

plt.title("Comparison of Property Tax Rate and Average School Rating")
plt.show()

# ----- Pairplot for Correlation Analysis -----
print("\nDisplaying Pairplot for Correlation Analysis")
selected_columns = ["numOfBedrooms", "numOfBathrooms", "propertyTaxRate", "avgSchoolRating"]
sns.pairplot(df_cleaned[selected_columns])
plt.show()

# ----- Correlation Heatmap -----
print("\nDisplaying Correlation Heatmap")

plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(df_cleaned[selected_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")

# Adding annotation inside the plot
plt.text(0.5, -0.5, "Darker colors indicate stronger correlations", fontsize=12, ha="center", va="center", transform=plt.gca().transAxes)
plt.show()

# ----- Scatter Plot for Highly Correlated Variables -----
print("\nDisplaying Scatter Plot: Bedrooms vs. Bathrooms")

plt.figure(figsize=(8, 5))
scatter = sns.scatterplot(x=df_cleaned["numOfBedrooms"], y=df_cleaned["numOfBathrooms"], alpha=0.6)
plt.xlabel("Number of Bedrooms")
plt.ylabel("Number of Bathrooms")
plt.title("Scatter Plot: Bedrooms vs. Bathrooms")

# Adding annotation inside the scatter plot
plt.text(2, 5, "Larger homes tend to have more bathrooms", fontsize=12, color="red")
plt.show()
```
